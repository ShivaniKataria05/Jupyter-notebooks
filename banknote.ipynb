```python
## Ensemble techniques 
```

```python
## Bagging (random forest)

from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(train_x,train_y)
predrf=rf.predict(test_x)
print(accuracy_score(test_y,predrf))
print(confusion_matrix(test_y,predrf))
print(classification_report(test_y,predrf))
```

```python
## boosting (adaboost)

from sklearn.ensemble import AdaBoostClassifier

## (base_estimator=DTC() , n_estimators=50 , learning_rate=1.0) --by default

ad=AdaBoostClassifier()
ad.fit(train_x,train_y)
predad=ad.predict(test_x)
print(accuracy_score(test_y,predad))
print(confusion_matrix(test_y,predad))
print(classification_report(test_y,predad))
```

```python
ad=AdaBoostClassifier(n_estimators=100)
ad.fit(train_x,train_y)
ad_pred=ad.predict(test_x)
print(accuracy_score(test_y,ad_pred))
print(confusion_matrix(test_y,ad_pred))
print(classification_report(test_y,ad_pred))
```

```python
from sklearn.svm import SVC
svc=SVC()

ad=AdaBoostClassifier(n_estimators=50 , base_estimator=svc,algorithm='SAMME')

ad.fit(train_x,train_y)
ad.pred=ad.predict(test_x)
print(accuracy_score(test_y,ad.pred))
print(confusion_matrix(test_y,ad.pred))
print(classification_report(test_y,ad.pred))
```

```python
svc=SVC(probability=True , kernel='linear')

ad=AdaBoostClassifier(n_estimators=50 , base_estimator=svc)
ad.fit(train_x,train_y)
ad.pred=ad.predict=(test_x)
#print(accuracy_score(test_y,ad.pred))
print(confusion_matrix(test_y,ad.pred))
print(classification_report(test_y,ad.pred))
```

```python
## Hyperparameter tuning -- GridsearchCV


from sklearn import svm
from sklearn.model_selection import GridSearchCV
svc=svm.SVC()

parameters= {'kernel':['linear','rbf'],'C':[1,10]}

clf = GridSearchCV(svc,parameters)
clf.fit(train_x,train_y)

print(clf)
print(clf.best_params_)
```

```python
svc=SVC()

para={'kernel':['linear','poly','rbf'],'C':[1,10]}

clf=GridSearchCV(svc,para)
clf.fit(train_x,train_y)

clf
```

```python
print(clf.best_params_)
```

```python
sv=svm.SVC(kernel='linear',C=1)

sv.fit(train_x,train_y)
t=sv.score(train_x,train_y)
print(t)
print(round(t,2))
```

```python
dtc=DecisionTreeClassifier()
param={'criterion':['gini','entropy']}

test=GridSearchCV(dtc,param)
test.fit(train_x,train_y)

```
